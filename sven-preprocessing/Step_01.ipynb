```python
import os
import pandas as pd
```

```python
directory = "./data/raw/"
```

```python
%%time
lstSensor = []
lstTouch = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    f = os.path.join(directory, filename)
    if "-points-" in filename:
        print(f)
        df = pd.read_csv(f, sep=";")
        lst = []
        for x in df.columns:
            lst.append(x.title().replace("-", ""))
        df.columns = lst
        x = f.split("-")
        df["PId"] = x[1]
        df["Phone"] = x[3]
        df.PId = df.PId.astype(int)
        df = df.drop_duplicates(keep="last")
        lstTouch.append(df)
    elif "-fitts-" in filename:
        continue
    elif f.endswith(".csv"):
        
        df = pd.read_csv(f, sep=";")
        lst =  []
        for x in df.columns:
            lst.append(x.title())
        df.columns = lst
        x = f.split("-")
        df["PId"] = x[1]
        df["Sensor"] = x[2]
        df["Phone"] = x[3]
        df.PId = df.PId.astype(int)
        df = df.drop_duplicates(subset=["Time"], keep="last")
        df = df.sort_values("Time")
        df = df.drop_duplicates(subset=["Time"], keep="last")
        lstSensor.append(df)
```

```python
dfSensors = pd.concat(lstSensor)
```

```python
dfTouch = pd.concat(lstTouch)
```
