```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import os.path
```

```python
participants = list(range(1, 21))
smartphones = ["S3Mini", "S4", "N5X", "N6"]
# TODO remove
participants = list(range(1, 5))
extension = 0

sensors = [
    "acc",
    "gyro",
    "ori",
    "grav",
    "mag",
    "rot"
]

tasks = [
    "points",
    "fitts"
]

file_names = sensors + tasks
```

```python
mean = dict()
for smartphone in smartphones:
    mean[smartphone] = dict()
    for sensor in sensors:
        mean[smartphone][sensor] = list()
```

```python
duration = dict()
for smartphone in smartphones:
    duration[smartphone] = list()
```

```python
for PID in participants:
    for smartphone in smartphones:
        print("processing - participant:", str(PID), "smartphone:", smartphone)
        # Read Files
        raw_data = dict()

        for file in file_names:
            file_path = str(Path.home()) + "/data/raw/fapra_imu-" +  str(PID) + "-" + file + "-" + smartphone + "-" + str(extension) + ".csv"
            if not os.path.isfile(file_path):
                print(file_path + "not found")
                continue
            raw_data[file] = pd.read_csv(file_path, ";")
        len(raw_data["points"])
        dur = (raw_data["points"].time.tolist()[-1] - raw_data["fitts"].time[0]) / 60000
        print(str(dur) + " min")
        duration[smartphone].append(dur)
        
        for sensor in sensors:
            raw_data[sensor] = raw_data[sensor].drop_duplicates(subset="time", keep="last")
            print(sensor, raw_data[sensor].diff().mean()["time"])
            mean[smartphone][sensor].append(raw_data[sensor].diff().mean()["time"])
```

```python
for smartphone in smartphones:
    for sensor in sensors:
        print(smartphone, sensor, sum(mean[smartphone][sensor]) / len(mean[smartphone][sensor]), "ms")
for sensor in sensors:
    tmp = list()
    for smartphone in smartphones:
        tmp.append(sum(mean[smartphone][sensor]) / len(mean[smartphone][sensor]))
    print("overall", sensor, sum(tmp) / len(tmp), "ms")
```

```python
tmp = []
for smartphone in smartphones:
    average_dur = sum(duration[smartphone]) / len(duration[smartphone])
    print(smartphone, average_dur, "min")
    tmp.append(average_dur)
print("overall", sum(tmp) / len(tmp), "min")
```
